Leer el archivo y almacenarlo en un DataFrame:

```python
import pandas as pd

# (a) Leer el archivo
college = pd.read_csv('College.csv')

```

Ajustar el índice y modificar la columna para usar los nombres de las universidades:

```python
# (b) Ajuste del índice
college2 = pd.read_csv('College.csv', index_col=0)
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')
college = college3  
```

Resumen numérico de los datos:

```python
# (c) Resumen numérico
resumen_numerico = college.describe()
print(resumen_numerico)
```

Matriz de dispersión entre las columnas Top10perc, Apps y Enroll:

```python
# (d) Matriz de dispersión
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']], alpha=0.2)

```

Boxplot para Outstate versus Private:

```python
import matplotlib.pyplot as plt

# (e) Boxplot de Outstate y Private
college.boxplot(column='Outstate', by='Private')
plt.title('Outstate Tuition vs Private')
plt.suptitle('')  
plt.show()
```

Crear una variable cualitativa Elite y hacer un conteo:

```python
# (f) Crear variable 'Elite' basada en Top10perc
college['Elite'] = pd.cut(college['Top10perc'], bins=[0, 50, 100], labels=['No', 'Yes'])


conteo_elite = college['Elite'].value_counts()
print(conteo_elite)


college.boxplot(column='Outstate', by='Elite')
plt.title('Outstate Tuition vs Elite')
plt.suptitle('')
plt.show()
```

Histograma para variables cuantitativas con distintos números de bins:

```python
# (g) Histogramas para variables seleccionadas
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

college['Apps'].plot.hist(bins=20, ax=axes[0, 0], title="Número de Aplicaciones")
college['Accept'].plot.hist(bins=20, ax=axes[0, 1], title="Número de Aceptaciones")
college['Enroll'].plot.hist(bins=20, ax=axes[1, 0], title="Número de Inscripciones")
college['F.Undergrad'].plot.hist(bins=20, ax=axes[1, 1], title="Estudiantes de Tiempo Completo")

plt.tight_layout()
plt.show()
```
